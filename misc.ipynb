```python
def load_set(file: h5py.File):
    sets = []
    for key in file.keys():
        raw_inputs   = torch.from_numpy(file[key]['data']['inputs'][()])
        lens_radius  = torch.full_like(input = raw_inputs,
                                       fill_value = file[key]['info']['lens_radius'][()])
        fs_distance  = torch.full_like(input = raw_inputs,
                                       fill_value = file[key]['info']['fs_distance'][()])
        lfl_distance = torch.full_like(input = raw_inputs,
                                       fill_value = file[key]['info']['lens_focal_length'][()])
        
        inputs       = torch.stack([raw_inputs, lens_radius, fs_distance, lfl_distance])
        outputs      = torch.from_numpy(file[key]['data']['outputs'][()]).unsqueeze(0)
        sets.append((inputs, outputs))
    return sets
```
